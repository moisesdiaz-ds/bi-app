```python
s3_client = boto3.client('s3')

def save_on_s3(x,name):
    csv_buffer = StringIO()
    x.to_csv(csv_buffer)
    csv_buffer.seek(0)
    s3_client.put_object(Bucket="mycomputer-csv", Key=f"model_evaluations/{eva_id}/{name}.csv", Body=csv_buffer.getvalue())
    return f"s3://mycomputer-csv/model_evaluations/{eva_id}/{name}.csv"
```

```python

def save_manifest_json(csv_file_location):
    # Initialize a session using AWS credentials
    aws_session = boto3.session.Session()

    # Initialize S3 client
    s3_client = aws_session.client('s3')

    # Define the location of your CSV file in S3 and where to save the manifest
    bucket_name = 'mycomputer-csv'  # Replace with your bucket name
    manifest_file_key = csv_file_location.split("mycomputer-csv/")[1].split(".")[-2]+"_manifest.json"  # Replace with your desired manifest file path

    # Create a manifest dictionary
    manifest = {
        "fileLocations": [
            {
                "URIs": [
                    csv_file_location
                ]
            }
        ],
        "globalUploadSettings": {
        "format": "CSV",
        "delimiter": ",",
        "containsHeader": True
      }
    }

    # Convert the manifest dictionary to a JSON string
    manifest_json = json.dumps(manifest, indent=2)

    # Upload the manifest JSON to the specified S3 bucket
    s3_client.put_object(Body=manifest_json, Bucket=bucket_name, Key=manifest_file_key)

    print(f'Manifest file uploaded to s3://{bucket_name}/{manifest_file_key}')
```

### Who are our good and bad customers? so the age, salary, work time + ratio

```python
data_payments = wr.s3.read_csv(f"s3://mycomputer-csv/model_evaluations/{eva_id}/data_payments.csv")
data_payslips = wr.s3.read_csv(f"s3://mycomputer-csv/model_evaluations/{eva_id}/data_payslips.csv")

data_payments = data_payments.drop(['id','user_id','created_at','full_name'],axis=1)
data_payslips = data_payslips.drop(['status','cancel_date','customer_id'],axis=1)
```

```python
# ### Aging

data_payments.paid_at = pd.to_datetime(data_payments.paid_at,format="%Y-%m-%d %H:%M:%S",errors='coerce')
data_payments['paid_at'] = pd.to_datetime(data_payments.paid_at.fillna('2050-01-12'))

aging_calculated = (pd.to_datetime(data_payments[data_payments.paid_at!='0000-00-00 00:00:00'].paid_at) - pd.to_datetime(data_payments[data_payments.paid_at!='0000-00-00 00:00:00'].finalized_at)).dt.days
data_payments['aging_calculated'] = 0
data_payments.loc[aging_calculated.index,'aging_calculated'] = aging_calculated.values

data_payments_filt = data_payments[data_payments.aging_calculated<=1000]

data_payments['aging'] = data_payments['aging_calculated']

# ### Aging plus
data_payments["aging_60plus"] = (
    data_payments["aging"] >= 60).astype(int)
data_payments["aging_30plus"] = (
    data_payments["aging"] >= 30).astype(int)
data_payments["aging_10plus"] = (
    data_payments["aging"] >= 10).astype(int)

data_payments["aging_15plus"] = (
    data_payments["aging"] >= 15).astype(int)

data_payments["aging_1plus"] = (
    data_payments["aging"] >= 1).astype(int)
```

```python
data_payslips.birth_date = pd.to_datetime(
    data_payslips.birth_date, errors='coerce')

# Convert the dates to datetime objects if they are not already
data_payslips['subscription_start_date'] = pd.to_datetime(data_payslips['subscription_start_date'], errors='coerce')
data_payslips['birth_date'] = pd.to_datetime(data_payslips['birth_date'])

# Calculate the difference in years
data_payslips['birth_date'] = pd.to_datetime(np.where(data_payslips['birth_date']<"1970-02-02","1970-02-02",data_payslips['birth_date'].astype(str)))
data_payslips['age'] = (data_payslips['subscription_start_date'] - data_payslips['birth_date']).dt.days/365


# ### Workplace time
data_payslips["date_service YYY.MM.DD"] = pd.to_datetime(data_payslips["date_service YYY.MM.DD"],format="%Y-%m-%d",errors='coerce')
data_payslips["workplace_time"] = (data_payslips.subscription_start_date - data_payslips["date_service YYY.MM.DD"]).dt.days/30
data_payslips["workplace_time"] = data_payslips["workplace_time"].fillna(0)
```

```python
data_payslips['age_cut'] = pd.cut(data_payslips['age'],[18,25,30,35,40,45,99])
data_payslips['salary_cut'] = pd.cut(data_payslips['salary'],[1000,1500,2000,3000,4000,6000,99999])
data_payslips['workplace_time_cut'] = pd.cut(data_payslips['workplace_time'],[0,50,100,200,500])
```

```python


data_payments_customer = data_payments.merge(
    data_payslips, left_on=["subscription_id"], right_on=["subscription_id"])

data_payments['finalized_at'] = pd.to_datetime(data_payments.finalized_at,format="%Y-%m-%d %H:%M:%S",errors='coerce')
data_payments_customer = data_payments_customer.sort_values(by=["subscription_id", "finalized_at"])
```
